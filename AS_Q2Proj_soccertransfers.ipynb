```python
#plot1
import plotly.express as px
# I USED AI FOR THIS ONE. I FED IT MY CODE AND TOLD TO ENHANCE
filtered_df = df[df['transfer_fee'] > 0]
fig = px.scatter(filtered_df,
                 x='market_value_in_eur',
                 y='transfer_fee',
                 size='transfer_fee',
                 color='transfer_season',
                 hover_name='player_name',
                 title='Transfer Fees vs. Market Value by Season',
                 labels={'market_value_in_eur': 'Market Value (EUR)', 'transfer_fee': 'Transfer Fee (EUR)'})
fig.update_layout(
    xaxis_title='Market Value (EUR)',
    yaxis_title='Transfer Fee (EUR)',
    legend_title='Season',
    template='plotly_white'
)
fig.show()
```

### description1
## Transfer Fees vs. Market Value by Season
This scatter plot shows the relationship between players' market value, which is how much they are worth, and their transfer fees, which is how much clubs paid for them. Each dot on the plot represents a player, and the size of the dot shows the amount of the transfer fee. The colors of the dots represent different seasons, making it easier to see how the data changes over time. From the plot, we can observe a few interesting trends. Some players had transfer fees much higher than their market value, suggesting that clubs might have thought these players were more valuable than their stats showed. Larger dots, which represent very high transfer fees, are rare, indicating that these expensive transfers don’t happen often. Finally, the dots don’t form a straight line, showing that the relationship between market value and transfer fees is not exact.


```python
#plot2
#  HALF AI AND HALF BRAIN USAGE FOR THIS ONE
import pandas as pd
import plotly.express as px
df = pd.read_csv('transfers.csv')
filtered_df = df[df['transfer_fee'] > 0]
avg_transfer_fee_by_club = filtered_df.groupby('to_club_name')['transfer_fee'].mean().sort_values(ascending=False).head(10)
fig1 = px.bar(avg_transfer_fee_by_club,
              x=avg_transfer_fee_by_club.index,
              y=avg_transfer_fee_by_club.values,
              labels={'x': 'Club', 'y': 'Average Transfer Fee (EUR)'},
              title='Top 10 Clubs by Average Transfer Fee',
              template='plotly_white')
fig1.show()
```

### description2
## Top 10 Clubs by Average Transfer Fee
This bar chart shows the top 10 clubs that spent the most money, on average, for their players. Each bar represents a club, and the height of the bar shows how much the club typically pays for a player. From the chart, we can see that some clubs spend much more than others, likely because they aim to buy the best players to strengthen their teams. The first few clubs have significantly higher bars, highlighting the big gap between the top spenders and the rest of the list. This chart helps us understand which clubs prioritize expensive transfers as part of their strategy to stay competitive.

```python
#plot3
#  HALF AI AND HALF BRAIN USAGE FOR THIS ONE
import pandas as pd
import plotly.express as px
df = pd.read_csv('transfers.csv')
df['month'] = pd.to_datetime(df['transfer_date']).dt.month
monthly_transfer_count = df['month'].value_counts().sort_index()
fig2 = px.line(x=monthly_transfer_count.index,
               y=monthly_transfer_count.values,
               labels={'x': 'Month', 'y': 'Number of Transfers'},
               title='Monthly Distribution of Transfers',
               template='plotly_white')
fig2.show()
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('transfers.csv')
df['transfer_date'] = pd.to_datetime(df['transfer_date'])
df['month'] = df['transfer_date'].dt.month
monthly_transfers = df.groupby('month').size()
plt.figure(figsize=(10, 6))
monthly_transfers.plot(kind='bar')
plt.title('Monthly Distribution of Transfers')
plt.xlabel('Month')
plt.ylabel('Number of Transfers')
plt.xticks(ticks=range(12), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=45)
plt.tight_layout()
plt.show()
```

### description3
## Monthly Distribution of Transfers
This plot shows the number of transfers that happen each month during the year. The x-axis represents the months, while the y-axis shows the number of transfers. From this chart, we can see that there are peaks in specific months, like July and August, which are likely during the summer transfer window. On the other hand, there are fewer transfers at the beginning and end of the year, probably due to league schedules. This pattern makes it clear that transfers are not spread out evenly but are concentrated in certain parts of the year.
